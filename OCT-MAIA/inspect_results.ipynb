```python
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
for filepath in results_dir_path.glob("**/*.pickle"):
    with filepath.open('rb') as file:
        preprocessed_oct_scans = pickle.load(file)
    
    print(preprocessed_oct_scans[0]._identifier.to_filename())
    n_scans = len(preprocessed_oct_scans)
    if n_scans > 0:
        fig, axs = plt.subplots(nrows=n_scans, ncols=1, figsize=(10, n_scans * 2))
        for i, scan in enumerate(preprocessed_oct_scans):
            if n_scans == 1:
                axs.imshow(scan.labelled_segmentation())
            else:
                axs[i].imshow(scan.labelled_segmentation())

        plt.show()
        input()
        clear_output(wait=True)
    else:
        print('No scans')
```

```python
n_scans = {'amd': 0, 'control': 0}
for filepath in results_dir_path.glob("**/*.pickle"):
    with filepath.open('rb') as file:
        preprocessed_oct_scans = pickle.load(file)
    for scan in preprocessed_oct_scans:
        if 'amd' in scan._identifier.to_filename():
            n_scans['amd'] += 1
        else:
            n_scans['control'] += 1
```
